```python
import matplotlib.pyplot as plt

companies = ['Company A', 'Company B', 'Company C', 'Company D']
market_share = [30, 25, 20, 25]

explode = [0.1 if share == max(market_share) else 0 for share in market_share]

plt.pie(market_share, labels=companies, explode=explode, autopct='%1.1f%%', 
        startangle=90, colors=['skyblue', 'lightgreen', 'orange', 'pink'])

plt.title("Market Share of Companies")
plt.show()
```
