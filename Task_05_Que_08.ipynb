```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = {
    'Math': np.random.randint(40, 100, 20),
    'Science': np.random.randint(35, 95, 20),
    'English': np.random.randint(45, 90, 20)
}

df = pd.DataFrame(data)

plt.boxplot([df['Math'], df['Science'], df['English']], labels=['Math', 'Science', 'English'])
plt.title("Student Exam Scores Distribution")
plt.ylabel("Scores")
plt.show()
```
