```python
import matplotlib.pyplot as plt
import numpy as np

months = np.arange(1, 13)
product_a = [5, 7, 8, 6, 7, 8, 9, 10, 8, 6, 7, 9]
product_b = [3, 4, 4, 5, 6, 5, 6, 7, 8, 7, 6, 5]
product_c = [2, 3, 2, 3, 4, 4, 3, 4, 5, 6, 5, 4]

plt.stackplot(months, product_a, product_b, product_c, 
              labels=['Product A', 'Product B', 'Product C'],
              colors=['skyblue', 'lightgreen', 'lightcoral'])

plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Cumulative Sales of 3 Products Over 12 Months")
plt.legend(loc='upper left')
plt.show()
```
